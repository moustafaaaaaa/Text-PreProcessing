# Convert To Lower Case

# Punctuation Removal

```python
def Punctuation_Removal(text) :
  punctuations = string.punctuation
  return text.translate(str.maketrans('','' , punctuations))
```

```python
x['clean_text'] = x['clean_text'].apply(lambda x : Punctuation_Removal(x))
```

# StopWords Removal

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

# stopwords removal function

def stopwords_removal(text) :
  return " ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
x['clean_text'] = x['clean_text'].apply(lambda x : stopwords_removal(x) )
```

```python
x.head()
```

# Removal of Frequent Words

```python
from collections import Counter
word_count = Counter()
for text in x['clean_text']:
  for word in text.split() :
    word_count[word] +=1


# Removal of Frequent Words function
frequent_words = set(word for (word , wc) in word_count.most_common(3))
def Removal_of_Frequent_Words(text) :
  return " ".join([word for word in text.split() if word not in frequent_words])
```

```python
word_count.most_common(10)
```

```python
frequent_words
```

```python
x['clean_text'] = x['clean_text'].apply(lambda x : Removal_of_Frequent_Words(x))
```

# Remove Rare Words

```python
rare_words = set(word for (word , wc) in word_count.most_common()[:-10:-1])

def remove_rare_words(text):
  return " ".join([word for word in text.split() if word not in rare_words])
```

```python
x['clean_text'] = x['clean_text'].apply(lambda x : remove_rare_words(x))
```

```python
x.head()
```

# Remove Regular Expression

```python
import re

def remove_special_char(text) :
  text = re.sub('[^a-zA-z0-9\s]' ,' ' , text)
  text = re.sub('\s+' , ' ' , text)
  return text.strip()
```

```python
x['clean_text'] = x['clean_text'].apply(lambda x : remove_special_char(x))
```

```python
x.sample(frac=0.5).head()
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()

def stemming(text) :
  return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
x['Stemmed Text'] = x['clean_text'].apply(lambda x : stemming(x))
```

# Lemmatization and pos tagging

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag , word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# pos tagging function

def get_wordnet_pos_tag(text) :

  if text.startswith('N') :
    return wordnet.NOUN

  elif text.startswith('J'):
    return wordnet.ADJ

  elif text.startswith('V'):
    return wordnet.VERB

  elif text.startswith('R'):
    return wordnet.ADV

  else :
    return wordnet.NOUN

# lemmatization function

def lemmatize_words(text) :
  lemmatizer = WordNetLemmatizer()
  tokinized_words = word_tokenize(text)
  pos_tagged = pos_tag(tokinized_words)

  return " ".join([lemmatizer.lemmatize(word , get_wordnet_pos_tag(pos)) for word , pos in pos_tagged])
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = 'hello iam desha i love Ai better going the best'
words = word_tokenize(text)
words_pos = pos_tag(words)
words_pos

lemmatizer = WordNetLemmatizer()
lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos_tag(pos))
        for word, pos in words_pos]
lemmatized_words
```

```python
x['lemmatized Text'] = x['clean_text'].apply(lambda x : lemmatize_words(x))
x.head()
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def spell_checking(text):
  corrected_words = []
  misspelled_words = spell.unknown(text.split())

  for word in text.split():
        if word in misspelled_words:
            correction = spell.correction(word)
            if correction is not None:
                corrected_words.append(correction)
            else:
                corrected_words.append(word)
        else:
            corrected_words.append(word)

  return " ".join(corrected_words)

```
